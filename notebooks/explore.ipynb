```python
import pandas as pd
table_names = [row.table_name for row in crsr.tables()]
tables = {}
for table_name in table_names:
    print(table_name)
    try:
        res = crsr.execute('SELECT * FROM {}'.format(table_name))
        columns = [column[0] for column in crsr.description]

        rows = res.fetchall()

        df = pd.DataFrame([list(r) for r in rows], columns=columns)

        tables[table_name] = df.set_index(df.columns[0])
    except:
        print('Error')
```
