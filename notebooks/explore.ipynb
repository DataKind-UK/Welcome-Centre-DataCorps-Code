```python
conn_str = (
    r'DRIVER={CData ODBC Driver for Access};'
    r'Data Source=ClientDatabaseStructure.mdb;'
    r'CHARSET=UTF16')
cnxn = pyodbc.connect(conn_str)
cnxn.setencoding('utf-16le')
cnxn.setdecoding(pyodbc.SQL_CHAR, encoding='utf-16le')
cnxn.setdecoding(pyodbc.SQL_WCHAR, encoding='utf-16le')
crsr = cnxn.cursor()
for row in crsr.tables():
    print(row)
```
